```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/xingxueyan/Desktop/Reviews_sub_10000.csv")
df1 = df.dropna()
```

```python
df.head()
```

```python
text = np.array(df1['Normalized_Text'])
```

```python
word_cloud = {}

for index,row in df1.iterrows():
    word_list = row["Normalized_Text"].split(" ")
    for i in word_list:
        if i in word_cloud.keys():
            word_cloud[i] += 1
        else:
            word_cloud[i] = 1
            
word_cloud

            
```

```python
keys = []
counts = []
for i in word_cloud.keys():
    keys.append(i)
    counts.append(word_cloud[i])

```

```python
data = pd.DataFrame(keys,counts)
```

```python
data.to_csv("counts")
```
